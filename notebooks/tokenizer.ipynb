```python
# Tokenizer Setup
Verifying the Tiktoken BPE encoding functionality and vocabulary bounds before integrating it into the main training loop.
```

```python
import tiktoken

# Initialize the standard GPT-2 tokenizer used for the NanoLLM
encoder = tiktoken.get_encoding("gpt2")
vocab_size = encoder.n_vocab

print(f"Vocabulary Size: {vocab_size}")

# Test Encoding
sample_text = "Building a 125M parameter model."
encoded = encoder.encode(sample_text, allowed_special={"<|endoftext|>"})
print(f"Encoded IDs: {encoded}")

# Test Decoding
decoded = encoder.decode(encoded)
print(f"Decoded Text: {decoded}")

# Verify tokenization matching
assert sample_text == decoded, "Decoding did not match original text."
```
